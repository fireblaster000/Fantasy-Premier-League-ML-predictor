```python
import requests
import pandas as pd

# Fetch FPL data
url = "https://fantasy.premierleague.com/api/bootstrap-static/"
response = requests.get(url)
data = response.json()

# Base player data
players_df = pd.DataFrame(data['elements'])

# Team and Position Mappings
teams_df = pd.DataFrame(data['teams'])[['id', 'name']].rename(columns={'id': 'team', 'name': 'team_name'})
positions_df = pd.DataFrame(data['element_types'])[['id', 'singular_name_short']].rename(columns={'id': 'element_type', 'singular_name_short': 'position'})

# Merge team names and positions
players_df = players_df.merge(teams_df, on='team', how='left')
players_df = players_df.merge(positions_df, on='element_type', how='left')

# Select final columns
important_columns = [
    "first_name", "second_name", "web_name", "team_name", "position",
    "now_cost", "total_points", "points_per_game", "minutes", "goals_scored",
    "assists", "clean_sheets", "goals_conceded", "yellow_cards", "red_cards",
    "penalties_saved", "penalties_missed", "saves", "bonus", "bps",
    "influence", "creativity", "threat", "ict_index", "selected_by_percent",
    "form", "ep_next", "expected_goals", "expected_assists","expected_goal_involvements",
    "expected_goals_conceded", "status", "penalties_order", "defensive_contribution", "defensive_contribution_per_90", "team_join_date"
]

filtered_df = players_df[important_columns].sort_values(by="total_points", ascending=False)

# Save to Excel
filtered_df.to_excel("fpl_players_24_25.xlsx", index=False)
print("Saved: fpl_players_24_25.xlsx ✅")
```

```python
import requests
import pandas as pd

# Fetch data from FPL API
fixtures_url = "https://fantasy.premierleague.com/api/fixtures/"
teams_url = "https://fantasy.premierleague.com/api/bootstrap-static/"

fixtures_data = requests.get(fixtures_url).json()
teams_data = requests.get(teams_url).json()['teams']

# Convert to DataFrames
fixtures_df = pd.DataFrame(fixtures_data)
teams_df = pd.DataFrame(teams_data)[['id', 'name']]
teams_df.columns = ['team_id', 'team_name']

# Map team names to fixture DataFrame
fixtures_df = fixtures_df.merge(teams_df, left_on='team_h', right_on='team_id', how='left')
fixtures_df = fixtures_df.rename(columns={'team_name': 'home_team'}).drop(columns=['team_id'])

fixtures_df = fixtures_df.merge(teams_df, left_on='team_a', right_on='team_id', how='left')
fixtures_df = fixtures_df.rename(columns={'team_name': 'away_team'}).drop(columns=['team_id'])

# Final columns
final_fixtures = fixtures_df[[
    'event', 'home_team', 'away_team', 'team_h_difficulty', 'team_a_difficulty'
]]
final_fixtures.columns = [
    'gameweek', 'home_team', 'away_team', 'home_team_difficulty', 'away_team_difficulty'
]

# Save to CSV
final_fixtures.to_excel("fpl_fixtures_named_24_25.xlsx", index=False)
print("excel saved: fpl_fixtures_named_24_25.xlsx")
```

```python
import requests
import pandas as pd

FPL_URL = "https://fantasy.premierleague.com/api/bootstrap-static/"

# --- Load bootstrap data ---
data = requests.get(FPL_URL).json()
players_df = pd.DataFrame(data["elements"])
teams_df = (
    pd.DataFrame(data["teams"])[["id", "name"]]
    .rename(columns={"id": "team_id", "name": "team_name"})
)

# --- Map team id -> team_name ---
players_df["team_id"] = players_df["team"]
players_df = players_df.merge(teams_df, on="team_id", how="left")

# --- Parse join date (keep timezone-aware to be safe) ---
players_df["team_join_date"] = pd.to_datetime(
    players_df["team_join_date"], errors="coerce", utc=True
)

# --- Convert selected_by_percent to float ---
players_df["selected_by_percent"] = pd.to_numeric(players_df["selected_by_percent"], errors="coerce")

# --- Season cutoff after 24/25 ended ---
season_cutoff = pd.Timestamp("2025-05-25", tz="UTC")

# --- Keep only AVAILABLE players who joined after cutoff ---
mask = (
    (players_df["status"] == "a") &
    (players_df["team_join_date"].notna()) &
    (players_df["team_join_date"] >= season_cutoff) &
    (players_df["minutes"] == 0) &  # Only those who haven't played yet
    (players_df["selected_by_percent"] > 3.0)  # Only those with >3% ownership
)

# --- Prepare output columns ---
pos_map = {1: "GKP", 2: "DEF", 3: "MID", 4: "FWD"}

cols = [
    "id", "first_name", "second_name", "web_name",
    "team_id", "team_name", "element_type",
    "now_cost", "selected_by_percent", "status", "team_join_date"
]

new_transfers = players_df.loc[mask, cols].copy()
new_transfers["position"] = new_transfers["element_type"].map(pos_map)

# --- Make date Excel-friendly ---
new_transfers["team_join_date"] = new_transfers["team_join_date"].dt.strftime("%Y-%m-%d")

# --- Sort and save ---
new_transfers = new_transfers.sort_values(["team_name", "second_name", "first_name"])
new_transfers.to_excel("new_transfers.xlsx", index=False)

print(f"Saved {len(new_transfers)} new transfers to new_transfers.xlsx")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

# --------------------------
# CONFIG
# --------------------------
EXCEL_PATH = "new_transfers.xlsx"  # Your input file
OUTPUT_PATH = "new_transfers_stats.xlsx"
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36"
}
SEASON = "2024"

# --------------------------
# HELPER FUNCTIONS
# --------------------------

def search_player_url(name):
    """Search Transfermarkt for the player and return the profile URL."""
    search_url = f"https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query={name.replace(' ', '+')}"
    print(f"🔍 Searching for {name}...")
    resp = requests.get(search_url, headers=HEADERS)
    soup = BeautifulSoup(resp.text, "html.parser")

    # Find first player result link
    link_tag = soup.select_one("table.items td.hauptlink a")
    if not link_tag:
        print(f"⚠️ No search result for {name}")
        return None
    profile_url = "https://www.transfermarkt.com" + link_tag.get("href")
    print(f"   ➡️ Found profile URL: {profile_url}")
    return profile_url


def get_player_id_slug(profile_url):
    """Extract slug and player ID from profile URL."""
    match = re.search(r"/([^/]+)/profil/spieler/(\d+)", profile_url)
    if not match:
        print("⚠️ Could not extract player slug and ID")
        return None, None
    return match.groups()


def detect_main_league_code(slug, player_id):
    """Find the domestic league competition code from the player's stats overview."""
    url = f"https://www.transfermarkt.com/{slug}/leistungsdaten/spieler/{player_id}/plus/1?saison={SEASON}"
    print(f"   🌍 Checking main league code from: {url}")
    resp = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(resp.text, "html.parser")

    # Look for competition link in the first table row
    comp_link = soup.select_one("table.items td.hauptlink a")
    if not comp_link or "wettbewerb" not in comp_link.get("href", ""):
        print("   ❌ Could not find competition link")
        return None

    match = re.search(r"wettbewerb/([A-Z0-9]+)", comp_link.get("href"))
    if not match:
        print("   ❌ Could not extract league code")
        return None
    league_code = match.group(1)
    print(f"   🏆 Detected main league code: {league_code}")
    return league_code


def scrape_league_stats(slug, player_id, league_code):
    """Scrape domestic league stats from Transfermarkt."""
    stats_url = f"https://www.transfermarkt.com/{slug}/leistungsdatendetails/spieler/{player_id}/plus/1?saison={SEASON}&verein=&liga=&wettbewerb={league_code}&pos=&trainer_id="
    print(f"   📊 Stats URL: {stats_url}")
    resp = requests.get(stats_url, headers=HEADERS)
    soup = BeautifulSoup(resp.text, "html.parser")

    table = soup.find("table", class_="items")
    if not table:
        print("   ❌ No stats table found")
        return None

    rows = table.find_all("tr", class_=["odd", "even"])
    if not rows:
        print("   ❌ No league rows found")
        return None

    cols = [col.get_text(strip=True) for col in rows[0].find_all("td")]
    if len(cols) < 13:
        print("   ⚠️ Unexpected table format")
        return None

    stats = {
        "matches": cols[4],
        "goals": cols[6],
        "assists": cols[7],
        "yellow_cards": cols[11],
        "red_cards": cols[12],
        "minutes": cols[-1]
    }
    print(f"   ✅ Stats found: {stats}")
    return stats


# --------------------------
# MAIN PROCESS
# --------------------------
df = pd.read_excel(EXCEL_PATH)

results = []

for idx, row in df.iterrows():
    full_name = f"{row['first_name']} {row['second_name']}"

    profile_url = search_player_url(full_name)
    if not profile_url:
        continue

    slug, player_id = get_player_id_slug(profile_url)
    if not slug or not player_id:
        continue

    league_code = detect_main_league_code(slug, player_id)
    if not league_code:
        continue

    stats = scrape_league_stats(slug, player_id, league_code)
    if not stats:
        continue

    result_row = {
        "first_name": row["first_name"],
        "second_name": row["second_name"],
        "league_code": league_code,
        **stats
    }
    results.append(result_row)

    time.sleep(1)  # be nice to the server

# Save results
if results:
    pd.DataFrame(results).to_excel(OUTPUT_PATH, index=False)
    print(f"\n💾 Saved stats for {len(results)} players to {OUTPUT_PATH}")
else:
    print("\n⚠️ No stats scraped.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
import unicodedata

# CONFIG
EXCEL_FILE = "new_transfers.xlsx"  # change to your file name
SEASON = 2024
OUTPUT_FILE = "player_stats.xlsx"

HEADERS = {
    "User-Agent": "Mozilla/5.0"
}

def normalize_name(name):
    return ''.join(
        c for c in unicodedata.normalize('NFKD', name)
        if not unicodedata.combining(c)
    )

def parse_card_string(card_str):
    """Convert '2\xa0/\xa0-\xa0/\xa0-' → yellow=2, second_yellow=0, red=0"""
    clean_str = card_str.replace("\xa0", "").strip()
    parts = [p.strip() for p in clean_str.split("/")]
    yellow = int(parts[0]) if parts[0] != "-" else 0
    second_yellow = int(parts[1]) if len(parts) > 1 and parts[1] != "-" else 0
    red = int(parts[2]) if len(parts) > 2 and parts[2] != "-" else 0
    return yellow, second_yellow, red

def search_player_url(name):
    """Search player profile URL on Transfermarkt"""
    search_query = normalize_name(name).replace(" ", "+")
    search_url = f"https://www.transfermarkt.com/schnellsuche/ergebnis/schnellsuche?query={search_query}"
    print(f"🔍 Searching for {search_url}...")
    resp = requests.get(search_url, headers=HEADERS)
    soup = BeautifulSoup(resp.text, "html.parser")
    profile_link = soup.select_one("table.items a[href*='/profil/spieler/']")
    if profile_link:
        return "https://www.transfermarkt.com" + profile_link["href"]
    return None

def get_main_league_code(profile_url):
    """Get the league code from the Leistungsdaten page"""
    player_id = re.search(r"/spieler/(\d+)", profile_url).group(1)
    stats_url = f"https://www.transfermarkt.com{profile_url.split('.com')[-1].replace('/profil', '/leistungsdaten')}/plus/1?saison={SEASON}"
    print(f"   🌍 Checking main league code from: {stats_url}")
    resp = requests.get(stats_url, headers=HEADERS)
    soup = BeautifulSoup(resp.text, "html.parser")
    league_link = soup.select_one("table.items a[href*='/wettbewerb/']")
    if league_link:
        match = re.search(r"/wettbewerb/([A-Z0-9]+)", league_link["href"])
        if match:
            return player_id, match.group(1)
    return player_id, None

def scrape_league_stats(slug, player_id, league_code, position):
    stats_url = f"https://www.transfermarkt.com/{slug}/leistungsdatendetails/spieler/{player_id}/plus/1?saison={SEASON}&wettbewerb={league_code}"
    print(f"   📊 Stats URL: {stats_url}")
    resp = requests.get(stats_url, headers=HEADERS)
    soup = BeautifulSoup(resp.text, "html.parser")
    table = soup.find("table", class_="items")
    if not table:
        print("   ❌ No stats table found")
        return None
    rows = table.find_all("tr", class_=["odd", "even"])
    if not rows:
        print("   ❌ No rows found in stats table")
        return None

    cols = [col.get_text(strip=True) for col in rows[0].find_all("td")]

    if position.lower() == "goalkeeper":
        if len(cols) < 9:
            print("   ⚠️ Unexpected column count for goalkeeper:", len(cols))
            return None
        yellow, second_yellow, red = parse_card_string(cols[5])
        stats = {
            "appearances": cols[3],
            "yellow_cards": yellow,
            "second_yellow": second_yellow,
            "red_cards": red,
            "goals_conceded": cols[6],
            "clean_sheets": cols[7],
            "minutes": cols[-1]
        }
    else:
        if len(cols) < 13:
            print("   ⚠️ Unexpected column count for outfield:", len(cols))
            return None
        stats = {
            "matches": cols[4],
            "goals": cols[6],
            "assists": cols[7],
            "yellow_cards": cols[11],
            "red_cards": cols[12],
            "minutes": cols[-1]
        }

    return stats

# MAIN SCRIPT
df = pd.read_excel(EXCEL_FILE)
results = []

for idx, row in df.head(10).iterrows():
    full_name = f"{row['first_name']} {row['second_name']}"
    position = row['position'] if pd.notna(row['position']) else ""
    print(f"\n🔍 Searching for {full_name} ({position})...")

    profile_url = search_player_url(full_name)
    if not profile_url:
        print("   ❌ Player profile not found")
        continue

    print(f"   ➡️ Found profile URL: {profile_url}")

    slug = profile_url.split("/")[3]
    player_id, league_code = get_main_league_code(profile_url)

    if not league_code:
        print("   ❌ League code not found")
        continue

    print(f"   🏆 Detected main league code: {league_code}")

    stats = scrape_league_stats(slug, player_id, league_code, position)
    if stats:
        results.append({"name": full_name, "position": position, **stats})

    time.sleep(2)  # polite delay

# Save results
if results:
    output_df = pd.DataFrame(results)
    output_df.to_excel(OUTPUT_FILE, index=False)
    print(f"\n✅ Saved stats for {len(results)} players to {OUTPUT_FILE}")
else:
    print("\n⚠️ No stats scraped.")
```
